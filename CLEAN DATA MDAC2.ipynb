```python
import pylab
import numpy as np
import pandas as pd
import seaborn as sn
from scipy import stats
from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../MDAC2/total16and17.csv')
```

```python
df.head()
```

```python
df = df.drop('Unnamed: 0',axis=1)
df.head()
```

```python
df['CustomerCounts'] = np.where((df['Covers2'] >= df['Checks2']),df['Covers2'],df['Checks2'])
df.head()
```

```python
df = df.drop(['Checks2','Covers2'],axis=1)
df.head()
```

```python
df = df.rename(columns = {'Gross Revenue2':'Revenue'})
df.head()
```

```python
df.dtypes
```

```python
df.Month.unique()
```

```python
# Month to Semester
# Jan, Feb, Mar, Apr --> Spring
# May, Jun, July --> Summer
# Aug, Sep, Oct, Nov --> Fall
# Dec --> Winter
conditions = [
    (df['Month'] == 'Jan'),
    (df['Month'] == 'Feb'),
    (df['Month'] == 'Mar'),
    (df['Month'] == 'Apr'), 
    (df['Month'] == 'May'),
    (df['Month'] == 'Jun'),
    (df['Month'] == 'July'),
    (df['Month'] == 'Dec')] 
choices = ['Spring', 'Spring','Spring','Spring','Summer','Summer','Summer','Winter']
df['Semester'] = np.select(conditions, choices, default='Fall')
df.Semester.unique()
```

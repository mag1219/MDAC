```python
# log(data+1)
df['log_Revenue'] = np.log(df['Revenue']+1)
df['log_CustomerCounts'] = np.log(df['CustomerCounts']+1)
df.head()
```

## One-hot Encoding 

```python
df.isnull().sum()
```

```python
# drop year column
df_new1 = df.drop(['Revenue','CustomerCounts','Year'],axis=1)
print(df_new1.shape)
df_new1.head()
```

```python
# create dummy variables for categorical features
df_new2 = pd.get_dummies(df_new1)
print(df_new2.shape)
df_new2.head()
```

## Correlation Matrix 

```python
print("Correlation Matrix")

def get_redundant_pairs(df_new2):
    '''Get diagonal and lower triangular pairs of correlation matrix'''
    pairs_to_drop = set()
    cols = df_new2.columns
    for i in range(0, df_new2.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop

def get_top_abs_correlations(df_new2, n=10):
    au_corr = df_new2.corr().abs().unstack()
    labels_to_drop = get_redundant_pairs(df_new2)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=False)
    return au_corr[0:n]

print("Top Absolute Correlations")
print(get_top_abs_correlations(df_new2, n=10))
```

```python
# drop columns which have strong correlation (corr >0.6) with each other
df_new3 = df_new2.drop(['Semester_Winter','Region_Restaurant','Profit Center_Profit Center Moreton Fig(287)'],axis=1)
print(df_new3.shape)
```

```python
# create train data for 1 revenue model and 2 customer counts model
train1 = df_new3.drop(['log_Revenue','log_CustomerCounts'], axis=1)
target1 = df_new3['log_Revenue']

train2 = df_new3.drop(['log_Revenue','log_CustomerCounts'], axis=1)
target2 = df_new3['log_CustomerCounts']

print(train1.shape, target1.shape, train2.shape, target2.shape)
```

## Regression Model for Revenue 

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.linear_model import RidgeCV
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.cross_validation import train_test_split
```

```python
# Split the dataset into training and test data
X_train, X_test, y_train, y_test = train_test_split(
    train1, target1, test_size=0.2, random_state=0)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
# y_test
```

```python
# Ridge Regression Model
print ("---Ridge Regression Model for Revenue---")
tuned_parameters = [{ 
                    'alpha' :[0.01, 0.1, 1, 10, 100]}
                ]
model1 = GridSearchCV(Ridge(), tuned_parameters, 
                       scoring='neg_mean_squared_error', cv=10, verbose=1
                     )
model1.fit(X_train, y_train)

print("Best parameters set found on development set:")
print(model1.best_params_)

print()
print("Grid scores on development set:")
for params, mean_score, scores in model1.grid_scores_:
    print("%0.3f (+/-%0.03f) for %r"
            % (mean_score, scores.std() * 2, params))
print()

y_hat = model1.predict(X_test)
mse = mean_squared_error(y_test, y_hat)
print("Mean Squared Error on test set = %0.4f" % mse)
```

```python
# rescaled y to original form
y_pred_rescaled = np.exp(y_hat)-1
y_test_rescaled = np.exp(y_test)-1
```

```python
# y_pred_rescaled
```

```python
df_hhh = y_test_rescaled.to_frame()
df_hhh.head()
```

```python
df_hhh['Predict_Revenue'] = y_pred_rescaled
df_hhh = df_hhh.rename(columns = {'log_Revenue':'True_Revenue'})
df_hhh.head()
```

```python
df_hhh.to_csv('train_revenue_result_ver4.csv', index=True)
```

## Regression Model for Customer Counts
